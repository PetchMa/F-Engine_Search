```python
from data_generation import create_data_set
import numpy as np
import gc
```

```python
plate1 = np.load('../../../../../../../datax/scratch/pma/real_filtered_HIP110750.npy')[20000:]
plate2 = np.load('../../../../../../../datax/scratch/pma/real_filtered_HIP13402.npy')[20000:]
plate3 = np.load('../../../../../../../datax/scratch/pma/real_filtered_HIP8497.npy')[20000:]
plate = np.vstack([plate1, plate2, plate3])
del plate1, plate2, plate3
gc.collect()
print(plate.shape)
```

```python
import time 
BATCH= 6
NUM_SAMPLES = 10000
for i in range(BATCH):
    print(i)
    start = time.time()
    data, false_data, true_data = create_data_set(plate, NUM_SAMPLES=NUM_SAMPLES, snr_base=20, snr_range = 10, factor=1)
    data_test, false_data_test, true_data_test = create_data_set(plate, NUM_SAMPLES=1000, snr_base=20, snr_range = 10, factor=1)
    print(time.time()-start)
    np.save('../../../../../../../datax/scratch/pma/TRAIN/data_'+str(i)+'.npy',data )
    np.save('../../../../../../../datax/scratch/pma/TRAIN/false_data_'+str(i)+'.npy',false_data )
    np.save('../../../../../../../datax/scratch/pma/TRAIN/true_data_'+str(i)+'.npy',true_data )
    
    np.save('../../../../../../../datax/scratch/pma/TRAIN/data_'+str(i)+'.npy',data_test )
    np.save('../../../../../../../datax/scratch/pma/TRAIN/false_data_test_'+str(i)+'.npy',false_data_test )
    np.save('../../../../../../../datax/scratch/pma/TRAIN/true_data_test_'+str(i)+'.npy',true_data_test )
    
    del data, false_data, true_data
    gc.collect()
```
