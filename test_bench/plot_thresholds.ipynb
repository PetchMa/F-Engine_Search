```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

false=[]
true=[]
true_single=[]
direct = ['VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.5.csv',
         'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.6.csv',
         'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.7.csv']


linestyles = ['-', '--', '-.', ':']
plt.figure(figsize=(15,10))
for i in range(3):
    df1 = pd.read_csv(direct[i])
    x = df1['SNR']
    false.append(df1['False'])
    true.append(df1['True'])
    true_single.append(df1['Single_True'])


    plt.plot(x, false[i], label = "False", linestyle=linestyles[i])
    plt.plot(x, true[i], label = "True", linestyle=linestyles[i])
    plt.plot(x, true_single[i], label = "True Single Shot",  linestyle=linestyles[i])

plt.xlabel('SNR')
# Set the y axis label of the current axis.
plt.ylabel('Accuracy ')
# Set a title of the current axes.
plt.title('Full Test 12000 Event 0.5-0.9 Threshold')
# show a legend on the plot
plt.legend()
plt.grid()
# Display a figure.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

false=[]
true=[]
true_single=[]
direct = ['VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.5.csv',
          'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.6.csv',
         'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.7.csv',
          'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.8.csv',
          'VAE-BLPC1-ENCODER_compressed_512v13-0onband_full_test_forest_0.9.csv']


# linestyles = ['-', '--', '-.', ':']
plt.figure(figsize=(15,10))
for i in range(len(direct)):
    df1 = pd.read_csv(direct[i])
    x = df1['SNR']
    false.append(df1['False'])
    true.append(df1['False'])
    true_single.append(df1['False'])

    overall = np.mean([false[i], true[i],true_single[i]],axis=0  )
    plt.plot(x, overall, label = "threshold "+str(0.5+i*0.1))

plt.xlabel('SNR')
# Set the y axis label of the current axis.
plt.ylabel('Accuracy ')
# Set a title of the current axes.
plt.title('Full Test Average 12000 Events 0.5-0.9 Threshold')
# show a legend on the plot
plt.legend()
plt.grid()
# Display a figure.
plt.show()
```
